```python
#Since this a seasonal data it isn't a stationary data and dataset should be stationary
#Applying dicky fuller test
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(series):
  result = adfuller(series)
  print('ADF Statistics : {}'.format(result[0]))
  print('p value : {}'.format(result[1]))
  if(result[1]<=0.05):
    print('Strong evidence against null hypothesis, reject the null hypothesis. Data has no unit root and is stationary')
  else:
    print('Weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary')
```

```python
adf_test(df['Thousands of Passengers'])
```

```python
#To make data stationary we will do differencing
df['Passengers First Difference']=df['Thousands of Passengers']-df['Thousands of Passengers'].shift(1)
```

```python
df.head()
```

```python
adf_test(df['Passengers First Difference'].dropna())
```

```python
df['Passengers Second Difference']=df['Passengers First Difference']-df['Passengers First Difference'].shift(1)
```

```python
df.head()
```

```python
adf_test(df['Passengers Second Difference'].dropna())
#Data has now become sationary
```

```python
#Since it is seasonal we will try to do it for 12.Since it is for entire year
df['Passengers 12 Difference']=df['Thousands of Passengers']-df['Thousands of Passengers'].shift(12)
```

```python
df.head()
```

```python
adf_test(df['Passengers 12 Difference'].dropna())
```

```python
from datetime import datetime,timedelta
```

```python
train_dataset_end=datetime(1955,12,1)
test_dataset_end=datetime(1960,12,1)
```

```python
train_data=df[:train_dataset_end]
test_data=df[train_dataset_end+timedelta(days=1):test_dataset_end]
```

```python
pred_end_date=test_data.index[-1]
pred_start_date=test_data.index[0]
```

```python
test_data
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model_ARIMA=ARIMA(train_data['Thousands of Passengers'],order=(10,2,0))#order=(p,q,d)
```

```python
model_ARIMA_fit=model_ARIMA.fit()
```

```python
model_ARIMA_fit.summary()
```

```python
pred=model_ARIMA_fit.predict(start=pred_start_date,end=pred_end_date)
residuals=test_data['Thousands of Passengers']-pred
```

```python
residuals
```

```python
model_ARIMA_fit.resid.plot(kind='kde')
```

```python
test_data['Predicted_ARIMA']=pred
```

```python
#ARIMA doesnt work well for seasonal data
test_data[['Thousands of Passengers','Predicted_ARIMA']].plot()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model_SARIMA=SARIMAX(train_data['Thousands of Passengers'],order=(3,0,5),seasonal_order=(0,1,0,12))
```

```python
model_SARIMA_fit=model_SARIMA.fit()
```

```python
model_SARIMA_fit.summary()
```

```python
pred_SARIMA=model_SARIMA_fit.predict(start=pred_start_date,end=pred_end_date)
residual_SARIMA=test_data['Thousands of Passengers']-pred_SARIMA
```

```python
model_SARIMA_fit.resid.plot(kind='kde')
```

```python
test_data['Predicted_SARIMAX']=pred_SARIMA
```
